```python
from pathlib import Path
import pandas as pd

# folder with the 14 downloaded files
src = Path("/Users/vlad/Desktop/fmri/group_project/voxel_wise_permutation/test_datasets/bp_sigma_batch")

# optional: save merged file in your SCIF folder
out = Path("/Users/vlad/Desktop/fmri/group_project/voxel_wise_permutation/datasets/vv_prm_n.csv")

def sigma_from_name(p: Path) -> float:
    # vv_prm_t_sigma_1p75.csv -> 1.75
    return float(p.stem.split("_")[-1].replace("p", "."))

files = sorted(src.glob("vv_prm_normal_sigma_*.csv"), key=sigma_from_name)
print("files found:", len(files))

df = pd.concat((pd.read_csv(f) for f in files), ignore_index=True)
#df = df.sort_values(["n", "sm_sigma", "snr"]).reset_index(drop=True)

print("rows:", len(df))
print("duplicates (n,snr,sm_sigma):", int(df.duplicated(["n", "snr", "sm_sigma"]).sum()))
print("unique sigmas:", sorted(df["sm_sigma"].unique()))

df.to_csv(out, index=False)
print("saved:", out)
```
